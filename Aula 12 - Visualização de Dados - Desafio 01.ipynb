### Importação da bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Mapear caminho e nome do arquivo
path = 'O:\\Curso Python- Big Data - Escola SENAI Suiço Brasileira\\Arquivos\\'
file = 'vendas_2.xlsx'

# Carregar o arquivo e inserir em um dataframe
df = pd.read_excel(path + file)

# Visualizar o cabeçalho do dataframe
df.head()
```

#### Criando um gráfico de colunas

```python
graficos = sns.barplot(x='Categoria',y='ValorVenda', data=df, orient='v', errorbar=None)

# O método sns.barplot vai retornar uma lista de sub
# métodos utilizando os containers para acessar o texto
# dos rótulos para cada coluna, desta forma o python entende
# que ele deve adicionar um rótulo em cada coluna, de acordo com o 
# comando graficos.containers[0]

graficos.bar_label(graficos.containers[0])
```

### Gráfico de linha Data x ValorVenda

```python
## Cria um novo dataframe utilizando as colunas Data Venda e ValorVenda
## e classifica o dataframe com base na coluna Data Venda

base = df[['Data Venda','ValorVenda']].sort_values('Data Venda')

# Agrupa dos dados por meio do campo Data e realiza a soma
base = base.groupby('Data Venda').sum()
```

```python
y = base['ValorVenda']
x = base.index

plt.plot(x,y)
```
