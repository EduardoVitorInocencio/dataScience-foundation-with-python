#### Exercício 01
Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.

```python
# Importar a biblioteca
import matplotlib.pyplot as plt

# Gerar os dados para inserir no gráfico
x = range(1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y (Três vezes X)')
print(y)

# Plotar um gráfico de linha

plt.plot(x,y) 

# Adicionar um título ao gráfico

plt.title('Draw a line', fontsize = 15, c='blue')

# Adicionar título aos eixos

plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
```

#### Exercício 04

Escreva um programa Python para desenhar gráficos de linhas dos dados financeiros da Alphabet Inc. entre 3 de outubro de 2016 a 7 de outubro de 2016. Exemplo de dados estão no arquivo fdata.csv. O trecho de código fornece a saída mostrada na captura de tela a seguir:

```python
# Importar as bibliotecas
import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho do documento
path = 'O:\\Curso Python- Big Data - Escola SENAI Suiço Brasileira\\Matplot\\\Lista 01\\'
file = 'fdata.csv'

# Abrir o documento
# O argumento index_col permite selecionar qual coluna será 
# o índice do dataframe exemplo: 0 - primeira coluna como índice
# 1 - segunda coluna como índice e None - padrão

df = pd.read_csv(path + file, sep = ',', index_col = None)

# Visualizar o cabeçalho
df.head()

# Plotar os dados
df.plot()
```

```python
df.head()
df.columns = ['data','abertura','alta','baixa','fechamento']
df.head()
```

```python
# Selecionando colunas do dataframe
df_2 = df[['data','abertura']]
df_2.info()
```

```python
# Substituir uma string

df_2['abertura'] = df_2['abertura'].replace('774. 25', '774.25')
df_2['abertura']

# Converter em Numérico pd.to_numeric(coluna)
df_2['abertura'] = pd.to_numeric(df_2['abertura'])
```

```python
x = df_2['data']
y = df_2['alta']

plt.plot(x,y)
```
