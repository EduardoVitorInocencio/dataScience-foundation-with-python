```python
# IMPORTAÇAO DAS BIBLIOTECAS

import pandas as pd #importar os dados ixternos (CSV)
import numpy as np # Algebra linear (vetores e arrays)
import matplotlib.pyplot as plt # Visualização de dados 
import seaborn as sns
from sklearn.linear_model import LinearRegression #Machine Learing, carregando o algoritmo de regressão
from sklearn.model_selection import train_test_split

caminho= './Arquivos/IceCreamData.csv'
dados= pd.read_csv(caminho)

dados.head()
dados.tail()
dados.describe()
dados.info()
```

# Criar e testar modelo

```python
#Criação da variável dependente/resposta (y):
y= dados['Revenue']

#Criação da variável independente/entrada (x)
X= dados[['Temperature']]

#Separar os dados para teste e treino 

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size= 0.25)
```

```python
modelo= LinearRegression()

modelo.fit(X_train,y_train)

print('Coef. linear (b):', modelo.intercept_)
print('Coef. angular (a):', modelo.coef_)
```

```python
#TESTANDO O MODELO 

y_predict = modelo.predict(X_test)
y_predict
```
