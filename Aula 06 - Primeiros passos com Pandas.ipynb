### Seção 01 - Trabalhando com Pandas SERIES

Uma série em Pandas é como uma coluna de uma tabela, ou seja, uma matriz de apenas uma coluna.

```python
# Utilizando o comando pd.Series()

idades = pd.Series([15,20,25,30,40,50,60])
print(idades)
```

```python
# Retornar apenas os valores da Série

idades.values
```

```python
# Retornar apenas os índices dos registros

idades.index
```

```python
# Alterando os índices

indice_novo = ['Ana','Amanda',
               'Jacqueline','Larissa',
               'Roberto','João','Alan']

idades_list = [15,20,25,30,40,50,60]

nova_serie = pd.Series(idades_list,index= indice_novo)

print(nova_serie)
```

### Acessando informações de uma Série

```python
print(nova_serie['Ana'])
```

### Seção 02: Trabalhando com DataFrame

Pandas Dataframe é uma estrutura de dados bidimensional, como uma matriz com múltiplas linhas e colunas.

```python
import pandas as pd

dados = {
    'calorias':[420,380,390],
    'duracao': [50,40,45]
}
# Carregando dados para um dataframe
df = pd.DataFrame(dados)

#print(df)

# Selecionando a coluna Calorias
print(df['calorias'])

# Selecionando a coluna Duração
print(df['duracao'])
```

```python
# Verificando apenas os valores do Dataframe

df.values
```

```python
# Verificando os índices do Dataframe

df.index
```

### Estatísticas básicas

A função df.describe() exibe as estatísticas básicas dos dados numéricos.

```python
df.describe()
```

### Estatísticas de dados categóricos (não numéricos)

Para estatísticas de variáveis categóricas, precisamos adicionar o argumento include = ['O'].

TOP: O valor que mais aparece nos campos

FREQ: Frequência (quantidade de vezes) que o valor mais comum aparece (TOP)

COUNT: Quantidade de elementos existentes nos campos

UNIQUE: Quantidade de elementos únicos (sem repetição) existentes nos campos

```python
df.describe(include=['O'])
```

### Estatísticas básicas de todos os dados

Ainda é possível aplicar o pd.describe a todos os dados. Para isso, precisamos adicionar o argumento include = 'all'.

```python
df.describe(include='all')
```
