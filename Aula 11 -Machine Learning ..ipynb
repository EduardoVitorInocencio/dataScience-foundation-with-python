#### Aplicao - Exemplo 01

```python
# IMPORTAAO DAS BIBLIOTECAS

import pandas as pd #importar os dados ixternos (CSV)
import numpy as np # Algebra linear (vetores e arrays)
import matplotlib.pyplot as plt # Visualizao de dados 

#Machine Learing, carregando o algoritmo de regresso 
from sklearn.linear_model import LinearRegression 

#Importando arquivos 
caminho= './Arquivos/cars.csv'

dados= pd.read_csv(caminho)

dados.shape

dados.head()

```

```python
#Excluir a primeira coluna do Dataframe
#Utilizando o comendo DF.DROP >> Axiz 1 (Coluna) >> Axois= 0 (linha)

dados= dados.drop(['Unnamed: 0'], axis=1)
```

```python
dados.head()
```

```python
# Separar as variveis em Independentes (X) e dependentes(y)

X= dados.loc[:,'dist'].values
y= dados.loc[:,'speed'].values

#print(X)
#print(y)
type(X)

#Clculo de correlao entre as variveis 
correlacao= np.corrcoef(X,y)
correlacao

#Formato de matriz com uma coluna a mais 
X= X.reshape(-1,1)

#Criao do modelo e treinamento
# (fit indica que o trinamento deve ser executado)

modelo= LinearRegression()
modelo.fit(X,y) #treinar com esses dados
```

```python
#Visualizao dos coeficientes 
modelo.intercept_
```

```python
#inclinao (angular)
modelo.coef_
```

### f(x)= ax+b --> 0,16556757x+ 8.283905641787172
#### a= 0,16556757
#### b= 8.283905641787172

```python
#Prever a velocidade para uma distancia de 22 metros
modelo.predict([[22]])
```

```python
# Visualizar os dados como grfico de disperso
plt.scatter(X,y)
plt.plot(X,modelo.predict(X),color='red')
```
