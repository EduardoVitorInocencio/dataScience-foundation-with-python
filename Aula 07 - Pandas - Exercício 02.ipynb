I) Gere um arquivo em Excel para representar as vendas efetuadas pela filial norte
em 2010.


```python
dados = df[(df['data']>='2010-01-01') & 
           (df['data']<='2010-12-31') & 
           (df['filial']=='Norte')
           ]

# Nome do arquivo
nome_arquivo = 'Norte_2010.xlsx'

# Salvar / Gerar um arquivo em excel
dados.to_excel(pasta_destino + nome_arquivo)
```

m) Gere um arquivo em Excel para representar as vendas efetuadas pela filial
Sudeste em 2010. Sendo que no relatório deve conter apenas vendas de casas
sem garagem.

```python
dados = df[(df['tipo']=='Casa') & 
   (df['garag.']==0) &
   (df['filial']=='Sudeste') &
   (df['data']>='2010-01-01') &
   (df['data']<= '2010-12-31')
   ]

# Nome do arquivo
nome_arquivo = 'Sudeste_2010_Casa_SemGarag.xlsx'

# Salvar / Gerar um arquivo em excel
dados.to_excel(pasta_destino + nome_arquivo)
```
