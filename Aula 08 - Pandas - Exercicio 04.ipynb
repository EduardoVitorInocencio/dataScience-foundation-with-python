Criando relatórios em função do arquivo dados-vendas.xlsx

a) Crie um dataframe, utilizando o arquivo dados-vendas.xlsx. Crie data frames
para dividir a planilha vendas em 4 arquivos. São eles:

- Tabela Fato

- Tabela Lojas

- Tabela Produtos

- Tabela Vendedor

```python
# Importar a biblioteca do Pandas

import pandas as pd

# Mapear o local do documento fonte

caminho = 'C:\\Users\\Aluno\\Documents\\Eduardo-Inocencio-Python\\Arquivos\\'
arquivo = 'dados-vendas.xlsx'

# Atribuir os dados deste arquivo, em um dataframe

base = pd.read_excel(caminho + arquivo, sheet_name='Vendas')

# Exibir os cabeçalhos das colunas

base.head()
```

```python
# Criando a tabela FATO com as colunas:

df_dados = base[['ID-Produto','Loja','ID-Vendedor','Data Venda','ValorVenda']]

# Renomear a coluna Loja para ID-Loja

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

df_dados.head()

# Salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'
arquivo = 'tabela_fato.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```

```python
# Criando a tabela FATO com as colunas:

df_dados = base[['ID-Produto','Produto','Categoria','Segmento','Fabricante']]

df_dados.head()

# Salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'
arquivo = 'tabela_produto.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```

```python
# Criando a tabela LOJAS com as colunas:

df_dados = base[['Loja','Cidade','Estado']]

# Renomear coluna

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

# Remover as duplicadas
df_dados = df_dados.drop_duplicates()

df_dados.head()

# Salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'
arquivo = 'tabela_lojas.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```
